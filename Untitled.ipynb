```python
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
import seaborn as sns
import numpy as np
import folium

%matplotlib inline
```

```python
df_clean = pd.read_csv('Chronic_dis.csv',index_col=0)
```

```python
df_clean.head()
```

```python
df2 = df_clean[(df_clean['Year'] != 2001)&(df_clean['Year'] != 2007)&(df_clean['Year'] != 2008)&(df_clean['Year'] != 2009)]
```

```python
df2.Year.value_counts()
```

```python
df_mortality = df2[(df2['QuestionID'] == 'OVC5_0')]
df_mortality.head()
```

```python
df_mortality.tail()
```

```python
df_mortality.shape
```

```python
usa1_map= folium.Map(location=[32.839681, -109.424421], zoom_start=4)

# Display only accidents where serious injuries where recorded
for lat, lng, label in zip(df_mortality.Latitute, df_mortality.Longitude, df_mortality.Topic):
    if label!='0':
        folium.Marker(
            [lat, lng] ).add_to(usa1_map)
    
# Show map
usa1_map
```
